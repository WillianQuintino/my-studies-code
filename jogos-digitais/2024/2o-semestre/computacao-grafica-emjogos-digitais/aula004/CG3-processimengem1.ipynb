```python
#ler e exibir uma imagem
img=cv2.imread('corgi-oculos-de-sol.jpg')
cv2_imshow(img)
```

```python
#dimenses da imagem
print(img.shape)
#print(img[0,0])

#acessando um pixel
b,g,r = img[0,0]
print(f'b=',b)
print(f'g=',g)
print(f'r=',r)
```
